# Asymptotic Cocentration of Area at Equators of Hyperspheres

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations
%matplotlib inline
```

```python
def plot_sphere(cmap, alpha=1):
    '''Takes a colour map and plots a 2-Sphere with that colour map. 
    You may also change the opacity via the alpha keyword 1 being the most opaque and 0 invisible.'''
    fig = plt.figure(figsize=(15,15))
    ax = fig.gca(projection='3d')

    ## Aesthetic alterations to the 3D plane

    # Limits the range of the axes
    ax.set_xlim((-1,1))
    ax.set_ylim((-1,1))
    ax.set_zlim((-1,1))

    # Hides the axis lines
    ax.w_xaxis.line.set_color((1.0, 1.0, 1.0, 0.0)) 
    ax.w_yaxis.line.set_color((1.0, 1.0, 1.0, 0.0)) 
    ax.w_zaxis.line.set_color((1.0, 1.0, 1.0, 0.0))

    # Removes the tick labels
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_zticklabels([])

    # Generate the points required for the sphere
    u, v = np.mgrid[0:2*np.pi:200j, 0:np.pi:200j]
    x = np.cos(u)*np.sin(v)
    y = np.sin(u)*np.sin(v)
    z = np.cos(v)
    ax.plot_surface(x, y, z, cmap = cmap, alpha = alpha)
```

```python
plot_sphere('inferno_r')
```

We may want to emphasise the fact that concentration occurs at the equators so we may consider using the shifted twilight colour map. (To see what colour maps available look at the following page: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html.)

```python
plot_sphere('twilight_shifted', 0.7)
```

When interpreting such results we must be careful. As noted by J. Matousek in his *Lectures on Discrete Geometry* (Springer Books, Vol 212, Chapter 14, Pg 299):
> Both measure concentration and the existence of almost-spherical sections are truly high dimensional phenomena, practically meaningless in the familiar dimensions 2 and 3. The low-dimensional intuition is of little use here, but perhaps by studying many results and examples one can develop intuition on what to expect in high dimensions.

When we say that concentration manifests at the equators, the fact we use *equators* instead of *equator* is important. By rotational symmetry we know that this phenomena cannot occur at a uniqe equator. The results really say that concentration manifests around *any* equator (which is simply the intersection of any hyperplane containing the origin and our sphere). As counterintuitive as this may sound, recall that this is a purely high dimensional phenomena and we require $n\rightarrow \infty$, thinking about a 3D sphere and its equators will be of no help.
